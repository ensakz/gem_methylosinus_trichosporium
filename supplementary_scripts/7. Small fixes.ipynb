## remove genes:

```python
empty_genes_to_remove = []
for gene in iMsOB3b.genes:
    if len(gene.reactions) == 0:
        empty_genes_to_remove.append(gene.id)
```

```python
empty_genes_to_remove
```

```python
cobra.manipulation.remove_genes(iMsOB3b, empty_genes_to_remove)
```

## remove metabolites:

```python
empty_metabolites_to_remove = []
for metabolite in iMsOB3b.metabolites:
    if len(metabolite.reactions) == 0:
        empty_metabolites_to_remove.append(metabolite.id)
```

```python
empty_metabolites_to_remove
```

```python
for metabolite in empty_metabolites_to_remove:
    iMsOB3b.metabolites.get_by_id(metabolite).remove_from_model()
```

```python
empty_metabolites_to_remove
```
